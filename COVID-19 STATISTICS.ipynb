```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("Latest Covid-19 India Status.csv")
```

```python
data.head(10)
```

```python
# Total states+union territories
data.shape
```

```python
# Information of the columns
data.info()
```

```python
# Getting list of columns present in the dataframe
data.columns
```

```python
# Replacing long name with short name
data=data.replace("Dadra and Nagar Haveli and Daman and Diu","Daman and Diu")
data.head(10)
```

```python
# Checking for any missing values
data.isnull().sum()
```

```python
# Barplot representing total cases in each state
plt.figure(figsize=(15,10))
sns.barplot(data=data,y="State/UTs",x="Total Cases")
```

This shows that Maharastra has maximum number of cases.

```python
# Death troll in each states.
plt.figure(figsize=(15,10))
sns.barplot(data=data,x="Deaths",y="State/UTs")
plt.show()
```

This again shows that Maharashtra suffered maximum number of deaths

# Q.1 Which are the top 3 states in accordance with total number of cases?

```python
data1=data.sort_values(by="Total Cases",ascending=False)
data1.head(3)
```

# Q.2 Which are the top 3 states in accordance with maximum number of deaths?

```python
data1=data.sort_values(by="Deaths",ascending=False)
data1.head(3)
```

# Q.3 Which states have high discharge rates?

```python
data1=data.sort_values(by="Discharge Ratio (%)",ascending=False)
data1.head()
```

# Q.4 Which state had maximum number of deaths?

```python
plt.figure(figsize=(15,10))
x=data["Death Ratio (%)"]
plt.pie(x,labels=data["State/UTs"],autopct="%.1f")
plt.legend(title="People Vs Sports",bbox_to_anchor=(1.5,0.5), loc="center right")
plt.show()
```

CONCLUSION: The death rate in Punjab is highest.

# Q.5 What is the relationship between the total number of cases and the number of people discharged?

```python
sns.relplot(x="Total Cases",y="Discharged",data=data)
plt.show()
```

CONCLUSION : As the number of cases increases , the number of recoveries also increases.
